# HTTP 통신 함수
### json 리턴 데이터를 검증용 CSV로 재변환

```python
def req_naver_price(req_data,sort):
    
    print(type(req_data))
    
    if sort == None:
        print('정렬이 비었음')
        sort = 'asc'
        
    http = httplib2.Http()
    print(headers)
    
    response,content = http.request(getUrl('asc',req_data['GOODS_NAME2']), 'GET', headers=headers)
    
    print('response : %d' % response.status)
    
    result_arr = []
    
    if response.status == 200:
        text = content.decode('utf-8')
        data = json.loads(text)
        
        print(data)
        
        result_arr.append(req_data['PRD_CD'])
        
        if len(data['items']) > 0:
            result_arr.append(data['items'][0]['title'])
            result_arr.append(data['items'][0]['hprice'])
            result_arr.append(data['items'][0]['lprice'])
            result_arr.append(data['items'][0]['link'])
        else:
            result_arr.append('no.title')
            result_arr.append('no.hprice')
            result_arr.append('no.lprice')
            result_arr.append('no.link')
    return result_arr
```

```python
len(csv_result)

temp_data = csv_result.loc[0:10]

# API 호출 결과를 담는 배열
temp_arr = []

for i in range(len(temp_data)):
    api_result = req_naver_price(csv_result.loc[i], None)
    
    temp_arr.append(api_result)
```

```python
temp_df = pd.DataFrame(temp_arr, columns=['PRD_CD','PRD_NM','HIGH','LOW','URL'])
```
